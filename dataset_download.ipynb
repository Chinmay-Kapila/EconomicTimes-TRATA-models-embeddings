```python
#--------7-----------------
import os
import json
import time
import requests
from datetime import datetime, timezone

# ==========================================================
# Configuration
# ==========================================================

SAVE_DIR = "./datasets"

BASE_DIR = os.path.join(SAVE_DIR, "CVE")
os.makedirs(BASE_DIR, exist_ok=True)

# Replace with your NEW NVD API key (or leave blank)
NVD_API_KEY = "YOUR_NEW_API_KEY"

HEADERS = {}
if NVD_API_KEY:
    HEADERS["apiKey"] = NVD_API_KEY

BASE_URL = "https://services.nvd.nist.gov/rest/json/cves/2.0"

# NVD recommends reasonable page sizes
RESULTS_PER_PAGE = 200

YEARS = [2023, 2024, 2025, 2026]

# ==========================================================
# Download CVEs
# ==========================================================

session = requests.Session()

for year in YEARS:

    print("\n" + "=" * 60)
    print(f"Downloading CVEs for {year}")
    print("=" * 60)

    start_date = f"{year}-01-01T00:00:00.000Z"

    current_year = datetime.now(timezone.utc).year

    if year == current_year:
        end_date = datetime.now(timezone.utc).strftime("%Y-%m-%dT%H:%M:%S.000Z")
    else:
        end_date = f"{year}-12-31T23:59:59.999Z"

    start_index = 0
    total_results = None

    while True:

        params = {
            "pubStartDate": start_date,
            "pubEndDate": end_date,
            "startIndex": start_index,
            "resultsPerPage": RESULTS_PER_PAGE
        }

        success = False

        # Retry up to 5 times
        for attempt in range(5):

            try:

                response = session.get(
                    BASE_URL,
                    params=params,
                    headers=HEADERS,
                    timeout=60
                )

                # Too many requests
                if response.status_code == 429:
                    wait = 10 * (attempt + 1)
                    print(f"Rate limited. Waiting {wait} seconds...")
                    time.sleep(wait)
                    continue

                response.raise_for_status()

                data = response.json()
                success = True
                break

            except requests.exceptions.HTTPError as e:
                print(f"\nHTTP Error: {e}")
                print("URL:", response.url)
                print("Response:")
                print(response.text)
                break

            except requests.exceptions.RequestException as e:
                wait = 5 * (attempt + 1)
                print(f"Network error: {e}")
                print(f"Retrying in {wait} seconds...")
                time.sleep(wait)

        if not success:
            print(f"Failed downloading {year} page starting at {start_index}")
            break

        if total_results is None:
            total_results = data["totalResults"]
            print(f"Total CVEs found: {total_results}")

        filename = os.path.join(
            BASE_DIR,
            f"{year}_{start_index}.json"
        )

        with open(filename, "w", encoding="utf-8") as f:
            json.dump(data, f, indent=2)

        print(
            f"Saved {filename} "
            f"({min(start_index + RESULTS_PER_PAGE, total_results)}/{total_results})"
        )

        start_index += RESULTS_PER_PAGE

        if start_index >= total_results:
            break

        # Respect API rate limits
        time.sleep(1)

print("\nAll downloads completed successfully!")
```

```python
import os

ROOT = "/content/drive/MyDrive/Cyber_Datasets"

def tree(path, prefix=""):
    items = sorted(os.listdir(path))

    for i, item in enumerate(items):
        full_path = os.path.join(path, item)
        connector = "└── " if i == len(items)-1 else "├── "

        print(prefix + connector + item)

        if os.path.isdir(full_path):
            extension = "    " if i == len(items)-1 else "│   "
            tree(full_path, prefix + extension)

print(os.path.basename(ROOT))
tree(ROOT)
```
